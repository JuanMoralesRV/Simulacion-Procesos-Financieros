```python
import numpy as np, pandas as pd, yfinance as yf, matplotlib.pyplot as plt
```

```python
ticker = 'SPY'
data = yf.download(ticker, start='2022-01-01', end='2025-10-15', progress=False)['Close']['SPY']
daily_returns = data.pct_change()
daily_returns = daily_returns.dropna()
sigma = daily_returns.std() * np.sqrt(252)
S0 = data.iloc[-1]
r = 0.0411
T = 100 / 365
 
num_simulations = 10000
Z = np.random.normal(0, 1, num_simulations)
St = S0 * np.exp( (r - 0.5 * sigma ** 2) * T + sigma * np.sqrt(T) * Z  )
K = 680
 
 
call = np.maximum(St - K, 0)
put = np.maximum(K - St, 0)
call_value = np.exp(-r * T) * np.mean(call)
put_value = np.exp(-r * T) * np.mean(put)
 
call_value
```
